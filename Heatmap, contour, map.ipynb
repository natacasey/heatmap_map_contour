```python
df = df.dropna(subset=['Team Against', 'Result 2','ID2'])
```

```python
NA1=df['Na 1']== 4
df= df[NA1]
df.head()
```

```python
df.dropna(subset=['Team Against',  'Result 2','ID2'])
```

```python
result =df['Result 2']== 'lost'
data= df[result]
data.head()
```

```python
data = data.rename(columns={'Result 2': 'Result_2','Team Against': 'Team'})
```

```python
data.head()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
fig, ax = plt.subplots(figsize=(14, 12))
# Basic 2D density plot
sns.set_style("white")
sns.kdeplot(df.X, df.Y)
#sns.plt.show()
 
# Custom it with the same argument as 1D density plot
sns.kdeplot(df.X, df.Y, cmap="Oranges")
sns.scatterplot(data=df, x="X", y="Y", color = 'dimgray', s = 100, alpha=0.4)
 
# Some features are characteristic of 2D: color palette and wether or not color the lowest range
#sns.kdeplot(df.X, df.Y,  shade=True, shade_lowest=True, )
#sns.plt.show()
plt.title('\nPASSES AND PERFORMANCE BY TOTTENHAM HOTSPUR\n', color='dimgray', loc='left', fontsize =18)
# Set x-axis label
plt.xlabel('\npass length (m)\n', fontsize = 14, color='dimgray')
# Set y-axis label
plt.ylabel('pass completion (%)\n', fontsize = 14, color='dimgray')


def label_point(x, y, val, val1, ax):
    a = pd.concat({'x': x, 'y': y, 'val':val, 'val1':val1 }, axis=1)
    for i, point in a.iterrows():
        ax.text(point['x']-.3, point['y']-0.5, str(point['val']+' '+'to'+ '\n'+point['val1']+'    '), fontsize=10.5, color = 'black')

label_point(data.X, data.Y, data.Result_2, data.Team, plt.gca())
ax.spines['left'].set_color('gray')
ax.spines['bottom'].set_color('gray')
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)
ax.tick_params(axis='x', labelsize=13) 
ax.tick_params(axis='y', labelsize=13) 
plt.show()
```

```python
latitude = df2['Latitude']
longitude = df2['Longitude']
```

```python
from mpl_toolkits.basemap import Basemap
fig, ax = plt.subplots(figsize=(20, 20))
map=Basemap(llcrnrlat=24,urcrnrlat=50,llcrnrlon=-130,urcrnrlon=-55,projection='cyl',resolution='c', suppress_ticks = True)
map.drawcoastlines()
map.drawcountries()
map.drawstates(color='dimgray')
plt.scatter(longitude, latitude, c='darkorange', alpha=0.7,s=130, edgecolors='lightgray')
plt.title('\nCOSTCO STORES IN THE CONTIGUOUS USA\n', color='dimgray', loc='left', fontsize =18)
ax.axis("off") 
plt.show()
```
